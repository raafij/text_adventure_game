```python
"""
Created on Tuesday, 13th Oct 2020
Submission on Saturday, 17th Oct 2020

@author: Raafi Jahangir
"""

"""
DocString: 
    
    Game based on Jumanji (the movie)

    A) Introduction:
    The user get to choose a character from Jumanji and several conflicts before they are put
    in a situation where they meet the boss and have to fight them.
    
    Round 1: Fighting off a hippo
    Round 2: Choosing between two paths
    Round 3: Final boss gun fight
    
    B) Known bugs and/or errors:
    None      
    
"""

#########################################################################################################  
#     Imported several Python packages to suit the game
#########################################################################################################

import sys
import time
import random as rand
    
#########################################################################################################  
#     Introduction function - asks if the user is ready to play and what their name is
#########################################################################################################

def intro():
    #Asks the user if they are ready or not
    ready = input(prompt = "\nRaafi: Are you ready? (y/n)\n\n> ").lower()
    
    print(f"""
    
******************************************************************************************

""")
    
    yes_options = ["y", "yes", "yep", "sure", "yeah", "of course","let's go!"]
    no_options = ["n", "no", "nah", "nope", "i don't want to play!"]
    
    #If the user chooses anything from the yes_options list - this will run
    if ready in yes_options:
        
        #Creating a global variable for user_name to recall throughout the game
        global user_name
        
        #Automatic capitalization of user name
        user_name = input(prompt = "Raafi: Got it! What is your name?\n\n> ").capitalize()
        
        #Their name has to have more than zero letter. Once TRUE - this will run
        if len(user_name) > 0:

            print(f"""
Raafi: Well, hello there, {user_name}! You now need to choose a character, here are your choices:""")
            
            #Moves user to the choose character stage
            return choose_character()
        
        #If the user did not type anything and wanted to skip - this will run
        else:
            print(f"""
Raafi: You didn't type anything. If you are not ready, I understand, I will give you time. 
""")
            return intro()

    #If the user chooses anything from the no_options list - this will run
    elif ready in no_options:
        print("\nRaafi: Okay I'll give you a few seconds to settle in.\n")
        time.sleep(3)
        return intro()
    
    #If the user doesn't type anything from the lists - this will run
    else:
        print("\nRaafi: Sorry, I didn't get that. Try again!\n")
        return intro()

#########################################################################################################
#     Defining the choose character stage where the user chooses one of the two characters
#########################################################################################################   

def choose_character():
    
    #Creating a global varible to recall throughout the game
    global character
    
    #User is able to choose from the two characters
    character = input(prompt = f"""
a) Dr. Xander Bravestone

b) Prof. Shelly Oberon


Who would you like to be?\n
{user_name}: """).lower()
    
    #Spacing
    print(f"""
        
******************************************************************************************
""")
    
    #Creating global varibles to recall throughout the game
    global xander_options
    global shelly_options
    
    #Creating a list for any possible user input in connection with character choice
    xander_options = ["a", "1", "one", "xander", "brave", "stone", "bravestone", "xander bravestone", "dr", "dr.", "doctor"]
    shelly_options = ["b", "2", "two", "shelly", "oberon", "shell", "shelly oberon", "prof", "prof.", "professor"]
    
    #If user chooses Xander - this will run
    if character in xander_options:
        return xander()
    
    #If user chooses Shelly - this will run
    elif character in shelly_options:
        return shelly()

    #If user chooses neither from the list - this will run
    else:
        print("\nRaafi: Sorry, I didn't get that. Please pick from the choices below:\n")
        return choose_character()    

#########################################################################################################    
#     Defining Xander function which goes into the Xander character and shows his statistics and allows
#     the user to type in their own strength and weakness
#########################################################################################################

def xander():        
    xander_strengths = ["Fearless", "Speed", "Climbing", "Smouldering Intensity", "Boomerang"]
    xander_weaknesses = ["Generiosity"]

    #Shows the user the character's stats
    print(f"""
Raafi:  Excellent choice! Here are your stats, Dr. {user_name} Bravestone!

        Character description:

        “The geekiest kid in school just became the biggest, buffest adventurer you've 
        ever seen, but he'll learn that true courage and confidence can't be programmed.”    

        Strengths:\t\t\t| Weaknesses:
        ________________________________|___________________________
\t\t\t\t\t|
        1) {xander_strengths[0]}\t\t\t| 1) {xander_weaknesses[0]}
        2) {xander_strengths[1]}\t\t\t|
        3) {xander_strengths[2]}\t\t\t|
        4) {xander_strengths[3]}\t|
        5) {xander_strengths[4]}\t\t\t|
""")

    global new_strength
    
    #Allows the user to type in their own strength
    new_strength = input(prompt = f"""
Raafi: Add a new strength to your list. What would you consider yourself to be good at?\n
{user_name}: """).capitalize()

    #Appends the new strength onto the strength list
    xander_strengths.append(new_strength)

    global new_weakness
    
    #Allows the user to type in their own weakness
    new_weakness = input(prompt = f"""
Raafi: Now, let's balance it with a new weakness. What would you consider yourself to be not so good at?\n
{user_name}: """).capitalize()

    #Appends the new weakness onto the weakness list
    xander_weaknesses.append(new_weakness)

    #Shows the user their updates stats
    print(f"""
Raafi:  Here are your updates stats:

        Strengths:\t\t\t| Weaknesses:
        ________________________________|___________________________
\t\t\t\t\t|
        1) {xander_strengths[0]}\t\t\t| 1) {xander_weaknesses[0]}
        2) {xander_strengths[1]}\t\t\t| 2) {xander_weaknesses[1]}
        3) {xander_strengths[2]}\t\t\t|
        4) {xander_strengths[3]}\t|
        5) {xander_strengths[4]}\t\t\t|
        6) {xander_strengths[5]}

        Let me save your information.
""")

    #Pause
    input("Press <Enter> to Save and Reload the game")
    return character_option()

#########################################################################################################    
#     Defining Xander function which goes into the Shelly character and shows her statistics and allows
#     the user to type in their own strength and weakness
#########################################################################################################

def shelly():
    shelly_strengths = ["Cartography", "Archeology", "Palaeontology"]
    shelly_weaknesses = ["Endurance"]

    print(f"""
Raafi:  Interesting choice! Here are your stats, Prof. {user_name} Oberon!

        Character description:

        “The coolest girl in school just got a harsh reality check. Trapped in the body 
        of an overweight, middle-aged man, her looks are no longer her greatest asset.”

        Strengths:\t\t\t| Weaknesses:
        ________________________________|___________________________
\t\t\t\t\t|
        1) {shelly_strengths[0]}\t\t\t|1) {shelly_weaknesses[0]}
        2) {shelly_strengths[1]}\t\t\t|
        3) {shelly_strengths[2]}\t\t|

""")
    
    global new_strength
    
    #Allows the user to type in their own strength
    new_strength = input(prompt = f"""
Raafi: Add a new strength to your list. What would you consider yourself to be good at?\n
{user_name}: """).capitalize()

    #Appends the new strength onto the strength list
    shelly_strengths.append(new_strength)
    
    global new_weakness
    
    #Allows the user to type in their own weakness
    new_weakness = input(prompt = f"""
Raafi: Now, let's balance it with a new weakness. What would you consider yourself to be not so good at?\n
{user_name}: """).capitalize()
    
    #Appends the new weakness onto the weakness list
    shelly_weaknesses.append(new_weakness)
    
    #Shows the user their updates stats
    print(f"""
Raafi:  Here are your updates stats:

        Strengths:\t\t\t| Weaknesses:
        ________________________________|___________________________
\t\t\t\t\t|
        1) {shelly_strengths[0]}\t\t\t| 1) {shelly_weaknesses[0]}
        2) {shelly_strengths[1]}\t\t\t| 2) {shelly_weaknesses[1]}
        3) {shelly_strengths[2]}\t\t| 
        4) {shelly_strengths[3]}

        Let me save your information.
""")

    #Pause
    input(prompt = "Press <Enter> to Save and Reload the game\n")
    return character_option()

#########################################################################################################    
#     Defining character option and creates variables to create user full name and partner full name
#########################################################################################################

def character_option():
    
    #static global variables to be able to pull throughout the game
    global character_full_name
    global partner_full_name
    
    print(f"""

******************************************************************************************

""")
    #if statement to determine character and partner's name depending on what user selected
    if character in xander_options:
        
        character_full_name = f"""Dr. {user_name} Bravestone"""
        partner_full_name = "Prof. Shelly Oberon"
        
        print(f"""Raafi: Welcome back, {character_full_name}

""")
        return level_one()

    else:
        character_full_name = f"""Prof. {user_name} Oberon"""
        partner_full_name = "Dr. Xander Bravestone"
        
        print(f"""Raafi: Welcome back, Prof. {user_name} Oberon!
        
""")
        return level_one()

#########################################################################################################    
#     Defining level one function where the user has to make a decision when a hippo appears out of the blue
#########################################################################################################

def level_one():
    
    input(prompt = """Press <Enter> to start""")
    
    print(f"""
******************************************************************************************""")
    
    print(f"""
Raafi: You and {partner_full_name} have just spawned in the middle of the jungle. 
       The first thing you have to do is.........
""")
    
    #Prints the next statement 2 seconds later
    time.sleep(2)
    
    print(f"""
Raafi: Look out.........
""")
    
    time.sleep(2)
    
    print(f"""
Raafi: There's a hippo rushing towards you! You have to think quick, what do you do?""")
    
    time.sleep(2)
    
    print(f"""
       a) Get into your fighting pose and get ready to attack!""")
    
    time.sleep(1)

    print(f"""
       b) Jump out of the way to dodge and then start running away.""")
        
    time.sleep(1)
    
    print(f"""
       c) Play dead. He might not be able to see you.""")
    
    time.sleep(1)

    answer = input(prompt=f"""
What do you pick?\n
{user_name}: """).lower()
    
    #Option list
    option_a = ["a", "1", "one", "fight", "attack"]
    option_b = ["b", "2", "two", "too", "jump", "dodge", "run", "run away", "jump out"]
    option_c = ["c", "3", "three", "tree", "play dead", "dead", "play"]

    if answer in option_a:
    
        print(f"""
Hippo: *rushing*""")

        time.sleep(2)

        print(f"""
{user_name}: *fighting pose* HIIII YAAA! *punch* *kick* *karate chop*""")

        time.sleep(2)

        print(f"""
Hippo: Ouch! *slowly limbs away* """)

        time.sleep(2)

        print(f"""
Raafi: Whoa, you brave person! I see that you used your "{new_strength.lower()}" strength which 
       allowed you to beat the hippo!
 
 
******************************************************************************************

""")
        
        input(prompt = "Press <Enter> to go to the next stage\n")
        return level_two()
        
    elif answer in option_b:
        
        time.sleep(1)
        
        print(f"""
Raafi: Good choice! You were able to outsmart the hippo and run away deeper into the jungle. 


******************************************************************************************

""")
        time.sleep(1)
    
        input(prompt = "\nPress <Enter> to go to the level\n")
        return level_two()
    
    elif answer in option_c:
        
        print(f"""
{user_name}: *closes eyes and stands still*""")

        time.sleep(2)

        print(f"""
Hippo: *rushing in and barges into you*""")

        time.sleep(2)

        print(f"""
{user_name}: Owwwww! *painful noises*""")

        time.sleep(2)

        print(f"""
*splash*""")

        time.sleep(2)
        
        print(f"""
Raafi: Unfortunately, the hippo attacks you point blank and throws you into the lake for the 
       crocodiles to eat you. Maybe this was because your "{new_weakness.lower()}" weakness showed up.""")

        #4 second delay so user is able to read and retry afterwards
        time.sleep(3)
        
        print(f"""
******************************************************************************************

                                    GAME OVER \U0001F44E

******************************************************************************************

""")
        
        #User loses - takes them to the fail function that allows user to replay the game
        return fail()

    else:
        print("\nRaafi: Sorry, I didn't get that. Restarting level.\n")
        return level_one()
    
#########################################################################################################
# Defining the fail function and allowing the user to decide whether or not they want to replay the game   
#########################################################################################################

def fail():
    replay = input(prompt = """
Raafi: Would you like to play again? (y/n)\n\n> """).lower()
    
    yes_options = ["y", "yes", "yep", "sure", "yeah", "of course", "let's go!"]
    no_options = ["n", "no", "nah", "nope", "i don't want to play!"]
    
    #If the user types anything from the yes_options list - it will restart the game
    if replay in yes_options:
        return intro()
    
    #If the user types anything from the no_options list - system will exit
    elif replay in no_options:
        print(f"""
Raafi: Thanks for playing, {user_name}!""")
        sys.exit(0)
    
    #If the user doesn't type from the following lists, it'll allow them to try the question again.
    else:
        print(f"""
Raafi: You didn't say yes or no. Try again.
""")
        return fail()
        
#########################################################################################################
#    Defining level two where the user has to make a decision on what path to take
#########################################################################################################

def level_two():
    
    print(f"""
Raafi: Congratulations for passing hippo, he sure looked like he wanted to eat you up. Onto the next.""")
    
    time.sleep(2)

    print(f"""
Raafi:  You come across two different paths. 

        Path 1 seems weirdly all clear. All the leaves are still. It doesn't look like anybody has 
        gone up this path in a long long time.
        
        Path 2 seems like there has been some activity going on. It looks like people's footsteps,
        but nothing in the far distance.""")
    
    time.sleep(4)

    answer = input(prompt=f"""
What do you pick? Path 1 or 2?\n
{user_name}: """).lower() 
    
    for element in answer:
        if "a" == element or "1" == element or "leaves" == element or "clear" in element \
        or "path 1" == element:
            
            time.sleep(1)
            
            print(f"""
{user_name}: *starts walking* """)   
            
            time.sleep(1)
            
            print(f"""
{partner_full_name}: AhhhhhH! *disappears*""")
            
            time.sleep(1)
            
            print(f"""
{user_name}: *rush to check on partner* *falls underneath*""")
            
            time.sleep(2)
        
            print(f"""
Raafi: Unfortunately, you and {partner_full_name} falls into a trap that looks like it had been set
       perfectly for outsiders. You are stuck and are unable to come out.""")

            #4 second delay so user is able to read and retry afterwards
            time.sleep(3)
            
            print(f"""
******************************************************************************************

                                    GAME OVER \U0001F44E

******************************************************************************************

""")
        
            #User loses - takes them to the fail function that allows user to replay the game
            return fail()
            
        if "b" == element or "2" == element or "activity" == element or "footsteps" in element \
        or "path 2" == element: 
            
            time.sleep(1)
            
            print(f"""
Raafi: Good choice - you see that the other path seemed too good to be true. You trust your instincts
       and follow the footsteps to lead to Van Pelt.
       

******************************************************************************************

""")
            time.sleep(1)
            
            input(prompt = "Press <Enter> to go to the next stage\n")
            return level_three()
            
        else:
            print("\nRaafi: Sorry, I didn't get that. Restarting level.\n")
            return level_two()
        
#########################################################################################################
#    Defining level three where the user finally finds the boss and has an epic showdown
#########################################################################################################

def level_three():
    
    time.sleep(2)
    
    print(f"""
Raafi: While you discover the depths of the jungle, you smell the fresh scent of fire. You sneak in
       the bushes and see Van Pelt cleaning his gun. What do you do?""")
    
    time.sleep(2)
    
    print(f"""
       a) Slowly turn around and walk away back into the jungle.""")
    
    time.sleep(1)

    print(f"""
       b) Peek in a little closer to see the surroundings.""")
        
    time.sleep(1)
    
    print(f"""
       c) Barge in guns blazing!""")
    
    time.sleep(1)

    answer = input(prompt=f"""
What do you pick?\n
{user_name}: """).lower()
    

    #For loop that allows the user to type any of these key words and the following function will run
    for element in answer:
        if "a" == element or "1" == element or "one" == element or "slow" in element or "turn" in element \
        or "walk" in element:
            
            time.sleep(1)
            
            print(f"""
Raafi:  You try turning around but accidentally step on a squirrel's tail. The squirrel runs away towards
        Van Pelt's direction. He is now aware that you are there.""")

            time.sleep(2)

            print(f"""
Van Pelt: *turns around to shoot you but misses*


******************************************************************************************

""")
    
            time.sleep(2)
            
            return accuracy_func()
    
        elif "b" == element or "2" == element or "two" == element or "too" == element or "peek" in element \
        or "closer" in element or "surroundings" in element:
        
            print(f"""
{user_name}: *Peeks*""")

            time.sleep(2)
        
            print(f"""
Raafi: A bird flies over you and screeches.""")

            time.sleep(2)
        
            print(f"""
Van Pelt: *Looks up* Hey! Who are you? Hands up. *aims gun*""")

            time.sleep(2)
        
            print(f"""
{user_name}: \U0001F632 \U0001F64C""")
        
            time.sleep(2)
        
            print(f"""
Raafi: Unfortunately you were caught and Van Pelt takes you into custody.""")
                    
            #4 second delay so user is able to read and retry afterwards
            time.sleep(3)

            print(f"""
******************************************************************************************

                                    GAME OVER \U0001F44E

******************************************************************************************

""")

            #User loses - takes them to the fail function that allows user to replay the game
            return fail()

        elif "c" == element or "3" == element or "three" == element or "tree" == element \
        or "barge" in element or "announce" in element or "fight" in element:
            
            print(f"""
{user_name} and {partner_full_name}: ARRRRRR! WE FOUND YOU!""")

            time.sleep(1)

            print(f"""
Van Pelt: *turns around to shoot you but misses*


******************************************************************************************

""")
    
            time.sleep(2)

            input(prompt = f"""
Raafi: Press <Enter> to continue with the final stage""")
            return accuracy_func()

        else:
            print("\nRaafi: Sorry, I didn't get that. Restarting level.\n")
            return level_three()        
    
#########################################################################################################
#     Defining the shooting function that simulates the shooting scenario where the user can choose 
#     between using a gun or sword. Using the gun leads to a battle with the boss which randomizes
#     the chances of the boss winning or losing. However this depends on the stats the user input
#     regarding their accuracy and boss difficulty.
#########################################################################################################

def shooting():
    
    print(f"""
Raafi:  You have two options!""")
    
    time.sleep(1)
    
    print(f"""
        a) Pull out your gun and shoot back""")

    time.sleep(1)

    print(f"""
        b) Draw your sword and rush to him""")

    time.sleep(1)

    action = input(prompt=f"""
What do you do?\n
{user_name}: """).lower()

    #Possible user inputs regarding choices
    option_a = ["a", "1", "one", "gun", "shoot", "shoot back"]
    option_b = ["b", "2", "two", "too", "sword", "rush", "run", "barge"]
    
    #If the user chooses anything from option_a - this will run
    if action in option_a:
        
        #Boss accuracy is randomized from 0 to what user input the boss difficulty to be.
        boss_accuracy = rand.randint(0,boss_difficulty_int)
        
        #If user accuracy is higher than boss accuracy - you win
        if accuracy_int > boss_accuracy:
            print(f"""
Van Pelt: *draws gun*""")               
            
            time.sleep(1)
        
            print(f"""  
{user_name}: *draws gun faster* *shoots Van Pelt right in the head*""")
            
            time.sleep(1)

            print(f"""  
Van Pelt: *gets shot* *falls dramatically in slow motion*""")
            
            time.sleep(3)
            
            return win()
        
        #If user accuracy is lower than boss accuracy - you lose. You get to restart the level
        elif accuracy_int < boss_accuracy:
            print(f"""
Van Pelt: *draws gun faster and shoots you*""")               
            
            time.sleep(1)
        
            print(f"""  
{user_name}: *gets shot* *falls dramatically in slow motion*""")
            
            time.sleep(1)

            print(f"""  
Raafi: Van Pelt was able to shoot you first and kill you. Better luck next time!""")
            
            print(f"""
******************************************************************************************

                                    GAME OVER \U0001F44E

******************************************************************************************""")
            return fail()

        else: 
            accuracy_int == boss_accuracy
            print("\nYour bullet hits each other!\n")
            input(prompt = "\nPress <Enter> to try again\n")
            return shooting()
                
    #If the user chooses anything from option_b - this will run
    elif action in option_b:
        
            time.sleep(1)
        
            print(f"""  
Van Pelt: *starts firing*""")
            
            time.sleep(1)
        
            print(f"""  
{user_name}: *gets shot* *falls dramatically in slow motion*""")
            
            time.sleep(2)
        
            print(f""" 
Raafi: Why would you ever bring a knife at a gun fight? You should know better. You are now dead.""")

            #4 second delay so user is able to read and retry afterwards
            time.sleep(3)
            
            print(f"""
******************************************************************************************

                                    GAME OVER \U0001F44E

******************************************************************************************

""")
        
            #User loses - takes them to the fail function that allows user to replay the game
            return fail()
    
    #If the user chooses anything apart from the options - this will run        
    else:
        print(f"""
Raafi:  Sorry, I didn't get that. Please choose from the choices again.
""")            
        input(prompt = "Press <Enter> to restart\n")
        shooting()
        
#########################################################################################################
#     Defining the accuracy function with allows the user to input their shooting accuracy. This will be 
#     taken into consideration when fighting with the boss and increases the chances to win if accuracy 
#     is higher
#########################################################################################################

def accuracy_func():
    
    accuracy_str = input(prompt = f"""
Raafi: How good do you rate your shooting accuracy (1-3)? (3 being best)\n\n{user_name}: """)
    
    global accuracy_int
    
    accuracy_1 = ["1", "one"]
    accuracy_2 = ["2", "two", "too"]
    accuracy_3 = ["3", "three", "tree"]
    
    #If function to change user inputs from strings into integer
    if accuracy_str in accuracy_1:                   
        accuracy_int = 1
        
    elif accuracy_str in accuracy_2:    
        accuracy_int = 2
    
    elif accuracy_str in accuracy_3:    
        accuracy_int = 3

    else:
        print(f"""
Raafi:  Sorry, I didn't get that. Please choose a number from 1-3.
""")            
        input(prompt = "Press <Enter> to restart\n")
        accuracy_func()
    
    return boss_accuracy_func()

#########################################################################################################
#     Defining the boss accuracy function with allows the user to input the boss difficulty.
#     This function take into consideration the user's accuracy and changes the value of boss difficulty
#     based on how good the user's accuracy is and how hard they want their boss to be
#########################################################################################################

        
def boss_accuracy_func():
        
    boss_difficulty_str = input(prompt = f"""
Raafi: Select your boss difficulty (1-3)? (3 being hardest)\n\n{user_name}: """)

    global boss_difficulty_int
    
    boss_difficulty_1 = ["1", "one"]
    boss_difficulty_2 = ["2", "two", "too"]
    boss_difficulty_3 = ["3", "three", "tree"]

    #If function that allows user to input boss difficulty and be affected by their own accuracy
    #The higher they choose their accuracy and lower they choose the boss difficulty, the easier
    #the level will be.
    if boss_difficulty_str in boss_difficulty_1:                   
        boss_difficulty_int = 4 - accuracy_int

    elif boss_difficulty_str in boss_difficulty_2:    
        boss_difficulty_int = 5 - accuracy_int

    elif boss_difficulty_str in boss_difficulty_3:    
        boss_difficulty_int = 6 - accuracy_int

    else:
        print(f"""
Raafi:  Sorry, I didn't get that. Please choose a number from 1-3.
""")            
        input(prompt = "Press <Enter> to restart\n")
        boss_accuracy_func()
    
    print(f"""
    
    
******************************************************************************************

""")
    
    input(prompt = f"""
Raafi: Press <Enter> to continue with the final stage""")
    return shooting()

#########################################################################################################
#     Defining the win funtion that gives a win message and exits out of the game
#########################################################################################################

def win():
    
    time.sleep(1)
    
    print(f"""
Raafi:  You beat Van Pelt! You were able to retrieve the jewel and saved the jungle from darkness and lifted
        its curse. Good job {user_name}. You give {partner_full_name} a huge hug (COVID rules don't apply in Jumanji).

        You win the game, thanks for playing! \U0001F4AA
        

******************************************************************************************
""")
    sys.exit(0)
            

#########################################################################################################
#     Defining the main funtion which gives the game context and starts the introduction function
#########################################################################################################

def main():
    
    #Game context printed out for the user to read
    print(f"""

Welcome to Jumanji! 

My name is Raafi Billingsley, Jumangi Field guide, I was hired by your former partner, now nemesis,
Professor Russel Van Pelt. He had an obsession with the Sacred Jewel of Jumanji.

After years of research, he set up an expedition to find the Jewel and took the trail to the Jaguar Shine.
Once he reached the destination, instead of protecting the Jewel, he took it out of the shrine and wore
it around his neck. In doing so, he put a curse on this jungle and darkness has ruled ever since.

Your job is to find him in this jungle and retrieve the Jewel.


   
******************************************************************************************  
    """)
    
    #Calls the intro function to start the introduction
    return intro()

#Calling the main function to start the game
main()
```
